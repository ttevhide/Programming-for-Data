1A.   Write a program that prompts the user to enter two floating point (double) numbers and an operator ( *, +, /, %, -). Display the results of the given operation.

```python
num1=input("Enter the first floating point number:")
num2=input("Enter the second floating point number:") 
num1=float(num1)
num2=float(num2) 
Operator=input("Enter the operator:") 
if Operator== "+":
  answer=(num1 + num2) 
elif Operator== "-":
  answer=(num1 - num2) 
elif Operator== "*":
  answer=(num1 * num2) 
elif Operator== "/":
  answer=(num1 / num2)
else:
  answer=(num1%num2) 
print(answer)
```

2A.   Write a method named isPrime, which takes an integer as an argument and returns true if the argument is a prime number, or false otherwise. Also write main method that displays prime numbers between 1 to 500.

```python
def isPrime(num): 
  prime=True
  for i in range(2,num): 
    if (num % i)==0:
      prime= False 
    if prime==True:
      return True
    else:
      return False

for i in range(1,501):
  if isPrime(i)==True:
    print(i)
```

2B.   Write a value-returning method, isVowel that returns the value true if a given character is a vowel, and otherwise returns false. In main() method accept a string from user and count number of vowels in that string.

```python
def isVowel(letter):
  if letter=="a"or letter=="A": 
    return True
  elif letter=="e" or letter=="E": 
    return True
  elif letter=="i" or letter=="I": 
    return True
  elif letter=="o" or letter=="O": 
    return True
  elif letter=="u" or letter=="U": 
    return True
  else:
    return False
string=input("Enter a string: ") 
count=0
for i in range(len(string)):
  if isVowel(string[i])==True: 
    count=count+1

print(count)
```

3B.   In a certain class, there are 5 tests worth 100 points each. Write a program that will take in the 5 tests scores from the user, store the tests scores in an array, and then calculate the students' average.

4B. A series of files have been sent to your program with different memory sizes. The sizes are all in whole numbers of Megabytes (MB). You have been asked to create a program that will count how many files there are, the total memory used and the average size of these files.

```python
def totals(array): 
  total=0
  for i in range(len(array)): 
    total=total+array[i]
  return total

store_space = [232,241,324,216,221,298,334,212,235,324,264,141]

print("The amount of files is ", (len(store_space))) 
print("The total memory used is ", totals(store_space)) 
print("The average memory used is ", (totals(store_space))/len(store_space))
```

5A.  You are a computing teacher and have to put the classes test scores into order(highest to lowest). The scores are entered into a variable as follows. 

scores=[66,52,74,88,62,98,72,48,82,58].

*   Attempt to put the scores into order(highest to lowest) using bubble sort then test to see if your program works. 
*   Then try to create another program which sorts the scores(highest to lowest) using insertion sort.




```python
scores=[66,52,74,88,62,98,72,48,82,58]

def bubblesort(lists): 
  length=(len(lists)) 
  swapped=True
  while swapped ==True: 
    swapped=False
    for i in range(length-1): 
      if lists[i]>lists[i+1]:
        new=lists[i] 
        lists[i]=lists[i+1]
        lists[i+1]=new 
        swapped=True
      length= length-1 
    return lists[:]
ordered= bubblesort(scores) 
print(ordered)


scores=[66,52,74,88,62,98,72,48,82,58]

def insertionsort(lists):
  for i in range(1, len(lists)): 
    value = lists[i]
    index = i-1
    while index >=0 and value < lists[index] : 
      lists[index+1] = lists[index]
      index -= 1 
    lists[index+1] = value
  return lists[:]


ordered= insertionsort(scores) 
print(ordered)
```
