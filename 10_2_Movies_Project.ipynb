### Clean the dataset

```python
df.drop_duplicates(inplace=True)
df.info()
df.head()
```

### How many movies of each genre are there?
---

```python
df['genre'].value_counts()
```

### Which director has the highest ranked movies?
---

```python
max_rank = df['rank'].max()
df[df['rank'] == max_rank]
```

### How many movies have ranks of over 9?
---


```python
len(df[df['rank'] > 9])
```

### Plot a bar chart of mean rank and genre
---

```python
import matplotlib.pyplot as plt

df2= df.groupby('genre')['rank'].mean()
  # Draw the bar graph
df2.plot(x='genre', y=df2, kind='bar')

plt.xlabel('Genre')
plt.ylabel('Rank')
plt.title('The Mean Rank')
  
plt.show()
```

### Plot a pie chart of how many movies of each genre there are 
---


```python
plt.pie(df2, labels=df2.keys())
plt.title('Movies by Genre')
plt.show()
```

### Plot a graph showing the mean Rank for each year 

```python
df3=df.groupby('year')['rank'].mean()
df3.plot(x='year', y=df3, kind='line')
plt.xlabel('Year')
plt.ylabel('Rank')
plt.title('The Mean Rank')
plt.show()
```
